# Analysis on FC Barcelona 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifaplayers = pd.read_csv('fifa_data.csv')
##filtering data for only fc barcelona players
Barcelona = fifaplayers.loc[fifaplayers.Club == 'FC Barcelona']
```

## Right and left-footed players

```python
#filtering data for left and right footed players
rightfoot = Barcelona.loc[Barcelona['Preferred Foot'] == 'Right'].count()[0]
leftfoot =  Barcelona.loc[Barcelona['Preferred Foot'] == 'Left'].count()[0]
```

```python
#pie chart viz on left and right foot players
plt.figure( figsize = (5,5))
labels = ['Left Footed', 'Right Footed']
plt.title('Right and Left-footed players', fontsize= 15)
plt.pie([rightfoot, leftfoot], labels = labels, autopct= '%.2f %%')
plt.show()
```

# Player ratings 

```python
Players = Barcelona['Name']
Ratings = Barcelona['Overall']
```

```python
plt.figure( figsize= (15,5))
plt.bar(Players, Ratings, width =0.9, color = '#b14e6a')
plt.title('FC Barcelona Player Ratings', color ='black', fontsize= 20)
plt.xticks(rotation=90, color= 'blue')
plt.yticks(np.arange(0,110, step =10))
plt.xlabel('Players', size =15)
plt.ylabel('Ratings', size =15)
    
plt.show()
```

# Player age grouping

```python
#filtering various age groups
young_players = Barcelona.loc[Barcelona.Age <= 20].count()[0]
mid_aged_players = Barcelona.loc[(Barcelona['Age'] > 20) & (Barcelona['Age'] <= 30)].count()[0]
old_players = Barcelona.loc[Barcelona['Age'] >= 31].count()[0]
```

```python
#plotting pie chart

plt.figure( figsize = (6,6))
labels = ['Players under age 20', 'Players between ages 20 & 30', 'Players above age 30']
plt.pie([young_players, mid_aged_players, old_players], labels = labels, autopct= '%.2f%%')
plt.title('Player Age Groups', fontsize= 15)

plt.show()
```
